# 1. Import Required Modules

```python
# Adjust working directory if needed
import os
os.chdir("path/to/traderzclient")

from accounts import Accounts
from orders import Orders
from pprint import pprint
```

# 2. Login and List Trading Accounts

```python
EMAIL = "example@gmail.com"
PASSWORD = "password"

# Login
acc = Accounts(email=EMAIL, password=PASSWORD)
print("Token:", acc.token)

# List trading accounts
print("\nAvailable trading accounts:")
accounts_list = acc.list_accounts()
for a in accounts_list:
    print(f"{a['tradingAccountId']} | {a['name']}")
```

# 3. Select a Trading Account

```python
# Select the second account for example
selected_id = accounts_list[1]["tradingAccountId"]
acc.select_account(trading_account_id=selected_id)

print("\nSelected account:")
pprint(acc.get_selected_account()["offer"]["name"])
```

# 4. Initialize Orders API

```python
orders_api = Orders(account=acc)
```

# 5. Retrieve Active Orders

```python
active_orders = orders_api.get_active_orders()
print("\nActive Orders:")
pprint(active_orders)
```

# 6. Create a New Pending Order

```python
new_order_id = orders_api.create_pending_order(
    instrument="EURUSD",
    order_side="BUY",
    volume=0.01,
    price=1.05,  # Limit price
    type="LIMIT",
    sl_price=0,
    tp_price=0
)

print("\nCreated Pending Order ID:", new_order_id)
```

# 7. Edit an Existing Pending Order

```python
updated_order_id = orders_api.edit_pending_order(
    instrument="EURUSD",
    order_id=new_order_id,
    order_side="BUY",
    volume=0.02,
    price_order=1.06,
    sl_price=1.04,
    tp_price=1.08
)

print("\nUpdated Pending Order ID:", updated_order_id)
```

# 8. Cancel a Pending Order

```python
cancelled_order_id = orders_api.cancel_pending_order(
    instrument="EURUSD",
    order_id=new_order_id,
    order_side="BUY"
)

print("\nCancelled Pending Order ID:", cancelled_order_id)
```
