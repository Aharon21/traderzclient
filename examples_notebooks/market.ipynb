# 1.Imports & Initialization

```python
import os
os.chdir("path/to/traderzclient")

from accounts import Accounts
from market import Market
from pprint import pprint
```

# 2. Login & List Trading Accounts

```python
# Login with your credentials
acc = Accounts(
    email="example@gmail.com",
    password="password"
)

print("Available trading accounts:")
for a in acc.list_accounts():
    print(f"- {a['tradingAccountId']} | {a['name']}")
```

# 3. Select a Trading Account

```python
# Automatically pick the first trading account
selected_id = acc.list_accounts()[0]["tradingAccountId"]

acc.select_account(selected_id)

print("Selected account:")
pprint(acc.get_selected_account())
```

# 4. Initialize the Market API

```python
market = Market(acc)
```

# 5. Retrieve All Available Symbols

```python
symbols = market.get_symbols()

print("Total symbols:", len(symbols))
print("First 5 symbols:")
for s in symbols[:5]:
    print("-", s["symbol"])
```

# 6. Market Watch â€” Get Live Quotes

```python
quotes = market.market_watch(["EURUSD", "XAUUSD"])

print("Market Watch Results:")
for q in quotes:
    print(f"{q['symbol']}: Bid={q['bid']} | Ask={q['ask']}")
```

# 7. Retrieve Account Balance

```python
balance = market.get_balance()

print("Account Balance:")
pprint(balance)
```

# 8. Retrieve Candlestick Data (OHLC)

```python
candles = market.get_candles(
    symbol="EURUSD",
    interval="M15",
    start="2025-10-14T00:00:00Z",
    end="2025-10-15T00:00:00Z"
)

print("First 3 candles:")
pprint(candles["candles"][:3])
```
